# Machine Anomaly Detection (MRI & X-Ray)
This notebook trains two separate Isolation Forest models to detect anomalous behaviors in MRI and X-Ray machine operational sensor data.

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import joblib
import os
os.makedirs('../models', exist_ok=True)

np.random.seed(42)
n_samples = 1000
```

## 1. MRI Machine Model
Sensors: Helium Level (%), Magnetic Field (T), RF Power (kW), Gradient Temperature (°C)

```python
# Normal operational data
normal_mri = pd.DataFrame({
    'helium_level': np.random.normal(loc=90, scale=3, size=n_samples),
    'magnetic_field': np.random.normal(loc=1.5, scale=0.01, size=n_samples),
    'rf_power': np.random.normal(loc=18, scale=1, size=n_samples),
    'gradient_temp': np.random.normal(loc=22, scale=1.5, size=n_samples)
})

# Anomalous operational data 
anomalous_mri = pd.DataFrame({
    'helium_level': np.random.uniform(low=30, high=60, size=50),
    'magnetic_field': np.random.uniform(low=1.3, high=1.7, size=50),
    'rf_power': np.random.uniform(low=22, high=30, size=50),
    'gradient_temp': np.random.uniform(low=30, high=45, size=50)
})

mri_data = pd.concat([normal_mri, anomalous_mri], ignore_index=True)

# Train & Save Isolation Forest
mri_model = IsolationForest(contamination=0.05, random_state=42)
mri_model.fit(mri_data)
joblib.dump(mri_model, '../models/mri_anomaly_model.pkl')
print("✅ MRI Anomaly Model trained and saved.")
```

## 2. X-Ray Machine Model
Sensors: Tube Current (mA), Exposure Time (ms), Voltage (kV), Detector Temp (°C)

```python
# Normal operational data
normal_xray = pd.DataFrame({
    'tube_current': np.random.normal(loc=300, scale=20, size=n_samples), # mA
    'exposure_time': np.random.normal(loc=150, scale=10, size=n_samples), # ms
    'voltage': np.random.normal(loc=120, scale=5, size=n_samples), # kV
    'detector_temp': np.random.normal(loc=25, scale=1.5, size=n_samples) # Celsius
})

# Anomalous operational data 
anomalous_xray = pd.DataFrame({
    'tube_current': np.random.uniform(low=100, high=500, size=50),
    'exposure_time': np.random.uniform(low=50, high=300, size=50),
    'voltage': np.random.uniform(low=80, high=160, size=50),
    'detector_temp': np.random.uniform(low=35, high=50, size=50)
})

xray_data = pd.concat([normal_xray, anomalous_xray], ignore_index=True)

# Train & Save Isolation Forest
xray_model = IsolationForest(contamination=0.05, random_state=42)
xray_model.fit(xray_data)
joblib.dump(xray_model, '../models/xray_anomaly_model.pkl')
print("✅ X-Ray Anomaly Model trained and saved.")
```
